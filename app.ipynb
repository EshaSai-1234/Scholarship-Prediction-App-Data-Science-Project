```python
X = data[["Years_at_Company","Satisfaction_Level","Average_Monthly_Hours"]]
y = data[["Salary"]]
```

```python
from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size= 0.2)
```

```python
X
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
```

```python
X_train
```

```python
import joblib
joblib.dump(scaler, "scaler.pkl")
```

```python
X_test = scaler.fit_transform(X_test)
```

```python
X_test
```

```python
X_test.shape
```

```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error

def results(predictions):
    print("Mean absolute error on model is {}".format(mean_absolute_error(y_test, predictions)))
    print("Root mean squared error on moel is {}".format(np.sqrt(mean_squared_error(y_test, predictions))))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
predictionslr = lr.predict(X_test)
```

```python
results(predictionslr)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
```

```python
svrmodel = SVR()
```

```python
param_gridsvr = {"C":[0.01,0.1,0.5], "kerner":["linear","rbf","poly"]}
```

```python
gridsvr = GridSearchCV(svrmodel, param_gridsvr)
```

```python
gridsvr.fit(X_train,y_train.values.ravel())
```

```python
gridsvr.best_params_
```

```python
predictionssvr = gridsvr.predict(X_test)
```

```python
results(predictionssvr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rfrmodel = RandomForestRegressor()
```

```python
param_gridfr = {"n_estimators": [2, 3, 4, 5, 6], "max_depth": [5, 10, 15]}
```

```python
gridfr = GridSearchCV(rfrmodel, param_gridfr)
```

```python
gridfr.fit(X_train, y_train.values.ravel())
```

```python
gridfr.best_params_
```

```python
predictionsofrfr = gridfr.predict(X_test)
```

```python
results(predictionsofrfr)
```
